```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/carlosfab/dsnp2/master/datasets/clustering.csv")
print("Tamanho do df: {}\n".format(df.shape))

df.head()
```

```

from sklearn.cluster import KMeans
```

```
# Instanciar e escolher os hiperparamemtros

model = KMeans(n_clusters=2)
model2 = KMeans(n_clusters=4)

# Separar os dados

X = df[['ApplicantIncome', 'LoanAmount']]

# Padronizar os dados de treino

#scaler = StandardScaler().fit(X)
#X = scaler.transform(X)

# Fit do modelo

model.fit(X)
model2.fit(X)

# Fazer previs√µes em cima dos novos dados

y_pred = model.predict(X)
y_pred2 = model2.predict(X)
```

```
plt.scatter('ApplicantIncome', 'LoanAmount', data=df)
plt.show()
```

```
plt.scatter('ApplicantIncome', 'LoanAmount', data=df, c=y_pred, s=40, cmap='viridis')
plt.show()
```

```
plt.scatter('ApplicantIncome','LoanAmount', data=df, c=y_pred2, s=40, cmap='viridis')
plt.show()
```
